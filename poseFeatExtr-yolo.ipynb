```lua
require 'paths'
require 'hdf5'
paths.dofile('util.lua')
paths.dofile('img.lua')
cutorch.setDevice(2)  -- using #1
require 'poseExtrUtil'
```

```lua
modelPath = 'umich-stacked-hourglass.t7'
batchSize = 5
inputFilePath = '/home/gengshan/workJul/darknet/results/comp4_det_test_person.txt'
outputFilePath = '/data/gengshan/pose/poseResult.h5'
```

```lua
-- Load pre-trained model
m = torch.load(modelPath)   

-- Init input buffer
centerList = {}
scaleList = {}
imgList = {}
imgBuff = torch.CudaTensor():resize(batchSize, 3, 256, 256)

-- Init output buffer
preds_hm={}
preds_img={}
hmBuff = torch.FloatTensor():resize(batchSize, 16, 64, 64)
outFile = hdf5.open(outputFilePath, 'w')

-- Load input
detList = readDectionList(inputFilePath, 10000, true)
detList['currPointer'] = 1
-- detList = readDectionList('tmpDetRes.txt')

-- Get result
beg = tonumber(os.date"%s")
collectgarbage()
collectgarbage()
lap = torch.zeros(9)
timer = torch.Timer()
-- while true do
for it = 1,1 do
    local remainBatch = table.getn(detList) - detList['currPointer'] + 1
    print(remainBatch .. ' left')
    if remainBatch == 0 then
        break
    end
    
    if remainBatch < batchSize then
        getBatch(remainBatch)
        getPred(remainBatch)     
        -- dumpResult(remainBatch)
    else    
        timerq = torch.Timer()   
        getBatch()
        lap[1] = lap[1] + timerq:time().real

        timerq = torch.Timer()     
        getPred()
        lap[2] = lap[2] + timerq:time().real

        -- dumpResult()

        if remainBatch % 50 == 0 then
            collectgarbage()
            collectgarbage()
        end
    end
end
print('cpu time lapsed ' .. timer:time().real .. 's ', 'with ')
print('time lapsed ' .. tonumber(os.date"%s") - beg .. 's ', 'with ')

print(lap[{{1, 2}}])
print(lap[{{3, 4}}])
print(lap[{{5, 9}}])

collectgarbage()
collectgarbage()
outFile:close()
```
